</t>What we are doing here is that we are going to take our test, train, and validation sets and decode them into an object in python using the json standard library in python.  We will read all the available fields except the orientation field to build an object.  With this new object in python we will then encode the objects into a COCO object in JSON format.</br>
</t>We should be able to test this technique by reading the newly created COCO objects and making sure that we return a null value when we query for the orientation field.  


**Implementation:**</br>


```python
# Import JSON Library
import json
import os


def recode(json_coco):
  for annotation in json_coco["annotations"]:
    annotation["bbox"] = annotation["bbox"][0:4]
  return json_coco
  


# Get a raw json file
# create a list of files 
fileList = os.listdir("./drive/MyDrive/mvtec_screws_v1.0")
# check that path is valid
# iterate over the directory
for fileName in fileList:
  
  # get a file object
  if fileName != "images":
    if fileName[len(fileName) - 4:] == "json":
      print("############################## " + fileName)
      file = open(("./drive/MyDrive/mvtec_screws_v1.0/" + fileName), 'r')

      # create json object
      json_coco = json.load(file)
      new_coco = recode(json_coco)
      print(new_coco["annotations"])
      new_coco = json.dumps(new_coco)
      newFile = open("new"+fileName,'w')
      newFile.write(new_coco)
  #write out to new cleaned json file  



```
